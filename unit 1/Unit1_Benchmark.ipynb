```python
from transformers import pipeline
```

# **TASK 1 : TEXT GENERATION**
## Text generation with BERT

```python
text_gen_bert = pipeline(
    "text-generation",
    model="bert-base-uncased"
)

text_gen_bert("The future of artificial intelligence is")
```

**Observation:**
BERT fails to generate fluent text or produces errors / nonsensical output.

**Explanation:**
BERT is an encoder-only transformer model. It predicts the missing tokens within a sentence by looking at both left and right context simultaneously.

## Text generation with RoBERTa

```python
text_gen_roberta = pipeline(
    "text-generation",
    model="roberta-base"
)

text_gen_roberta("The future of artificial intelligence is")
```

**Observation:**
RoBERTa also fails or outputs meaningless results when forced to generate text.

**Explanation:**
RoBERTa is architecturally identical to BERT and is also an encoder-only model. Although it is trained with more data and optimized hyperparameters, its training objective remains masked language modeling.
As a result, RoBERTa lacks the architectural capability to generate text sequentially, making it unsuitable for text generation tasks.

## Text generation with BART

```python
text_gen_bart = pipeline(
    "text-generation",
    model="facebook/bart-base"
)

text_gen_bart("The future of artificial intelligence is", max_length=40)
```

**Observation:**
BART generates fluent, coherent, and grammatically correct text.

**Explanation:**
BART is an encoder–decoder transformer model. The encoder processes the input text, while the decoder generates output token-by-token in an autoregressive manner.
This architecture is explicitly designed for generation tasks, which is why BART performs well at text generation, unlike encoder-only models.

# **TASK 2 : FILL MASK**

## Fill mask with BERT

```python
fill_mask_bert = pipeline(
    "fill-mask",
    model="bert-base-uncased"
)

fill_mask_bert("The goal of Generative AI is to [MASK] new content.")
```

**Observation:**
BERT predicts accurate and contextually appropriate words for the masked token.

**Explanation:**
Fill-mask prediction is BERT’s primary training objective. During pretraining, BERT learns to predict masked words using full bidirectional context.
This makes BERT highly effective at understanding sentence semantics and predicting missing tokens accurately.

## Fill mask with RoBERTa

```python
fill_mask_roberta = pipeline(
    "fill-mask",
    model="roberta-base"
)

fill_mask_roberta("The goal of Generative AI is to <mask> new content.")
```

**Observation:**
RoBERTa produces very accurate and confident mask predictions.

**Explanation:** RoBERTa improves upon BERT by using larger datasets, dynamic masking, and removing the Next Sentence Prediction objective.
Since it remains an encoder-only masked language model, it excels at fill-mask tasks and often performs even better than BERT.

## Fill mask with BART

```python
fill_mask_bart = pipeline(
    "fill-mask",
    model="facebook/bart-base"
)

fill_mask_bart("The goal of Generative AI is to <mask> new content.")
```

**Observation:**
BART fails or gives unexpected output for fill-mask prediction.

**Explanation:**
BART is not trained using a masked language modeling objective like BERT or RoBERTa. Instead, it is trained using sequence-to-sequence denoising, where entire spans of text are corrupted and then reconstructed.
As a result, BART does not naturally support single-token mask prediction, leading to poor or failed performance in fill-mask tasks.

# **TASK 3 : QUESTION ANSWERING**

## Question answering with BERT

```python
qa_bert = pipeline(
    "question-answering",
    model="bert-base-uncased"
)

qa_bert(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)
```

**Observation:**
BART fails or gives unexpected output for fill-mask prediction.

**Explanation:**
BART is not trained using a masked language modeling objective like BERT or RoBERTa. Instead, it is trained using sequence-to-sequence denoising, where entire spans of text are corrupted and then reconstructed.
As a result, BART does not naturally support single-token mask prediction, leading to poor or failed performance in fill-mask tasks.

## Question answering with RoBERTa

```python
qa_roberta = pipeline(
    "question-answering",
    model="roberta-base"
)

qa_roberta(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)
```

**Observation:**
RoBERTa produces accurate answers, often comparable to or better than BERT.

**Explanation:**
RoBERTa uses the same encoder-only architecture as BERT but benefits from improved pretraining strategies.
Its strong contextual understanding makes it highly effective for extractive question answering tasks.

## Question answering with BART

```python
qa_bart = pipeline(
    "question-answering",
    model="facebook/bart-base"
)

qa_bart(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)
```

**Observation:**
BART may produce weaker or inconsistent results compared to BERT and RoBERTa.

**Explanation:**
BART is optimized for generative tasks, not extractive question answering.
While it can understand the context, its decoder-based generation is less precise at identifying exact answer spans, which can reduce accuracy in extractive QA settings.
